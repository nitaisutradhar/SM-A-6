# Task 1: Basic Matplotlib (3 plots)

## 1. Bar chart of gym membership types
Use: gym['abonoment_type'].value_counts()

```python
plt.figure(figsize = (8,5))
membership_counts = gym['abonoment_type'].value_counts()
plt.bar(membership_counts.index, membership_counts.values)
plt.title('Distribution of Gym Membership Types')
plt.xlabel('Membership Types')
plt.ylabel('Number of members')
plt.show()
```

## 2. Histogram of time in gym
Use: plt.hist(gym['avg_time_in_gym'])

```python
plt.figure(figsize=(8,5))
plt.hist(gym['avg_time_in_gym'], bins=15, color = 'salmon', edgecolor="black")
plt.title('Distribution of Average Time Spent in Gym')
plt.xlabel('Time (Minutes)')
plt.ylabel('Frequency')
plt.show()
```

## 3. Scatter plot: Age vs visits per week
Use: plt.scatter(gym['Age'], gym['visit_per_week'])

```python
plt.figure(figsize=(8,5))
plt.scatter(gym['Age'], gym['visit_per_week'], alpha=0.5, color='purple')
plt.title("Memeber Age vs. Weekly Visit Frequency")
plt.xlabel("Age")
plt.ylabel("Visits per Week")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# Task 2: Basic Seaborn (3 Plots)

```python
# Set a consistent visual style
sns.set_theme(style="whitegrid")
```

## 1. Box plot: Gender vs Age
Use: sns.boxplot(data=gym, x='gender', y='Age')

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=gym, x= 'gender', y = 'Age', palette='Set2')
plt.title('Age Distribution by Gender')
```

## 2. Count plot: Days per week
Use: sns.countplot(data=gym, x='days_per_week')

```python
plt.figure(figsize=(8,6))
sns.countplot(data=gym, x = 'days_per_week', palette='viridis')
plt.title("Frequency of Gym Visits per Week")
```

## 3. Bar plot: Average visits by membership type
Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')

```python
plt.figure(figsize=(8,6))
sns.barplot(data=gym, x = 'abonoment_type', y = 'visit_per_week', palette='magma')
plt.title('Average Weekly Visits by Membership Type')
```
